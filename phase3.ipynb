# The College town factor: The effect colleges on housing prices in U.S. cities with populations less than 250,000

**College Town definition**: Throughout this study, a **college town** will be defined as a U.S. city that is home to at least one college or university, and has a population less than 285,000 people.

## Brief description of the data (review)
This study combines three sets of data: From Zillow.com, sale prices of single family homes in the United States since 1996. From the U.S. Census, population sizes of U.S cities since 2010. And from Wikipedia, a list of which U.S. cities have universities. 

The combined dataset contains 26,172 rows, each representing a city in the United States. Each row contains the following data to various levels of completeness.

### Zillow Data 
All 26,172 cities in the dataset contain data for their Mean single-family home sale price by month. Data points begin January 1996 and continue, one for every month, until September 2020 for a total of approximately 300 columns. Records begin at various starting months for different cities, but within each city, once data begins, it is complete from its starting point until the present. Zillow also assigns each city a size rank value, but not the absolute population.

### Population Data
16,966 cities contain Census Population data from 2010 to 2019 inclusive. The 2010 data for each of these cities represent an actual count whereas subsequent years are based on estimates.

### Colleges Data
780 cities are identified as having one or more colleges present.

## Research questions this study will answer

Initial data exploration revealed that since 1996, college towns have seen a higher average increase in both single family home sale prices, and in population, than have like-sized, non-college town cities. This study aims to broadly examine the qustion of whether college towns are better financial investments than non-college towns through answering specific research questions. These questions evaluate the return on investment in good years, the reduction of risk in down years, and the effects of extreme events such as the 2008 housing crash and subsiquent great recession, and the 2020 COVID epidemic.

## Preregistration of analyses: Specific analyses that will be performed
In years when the average sale price of single family homes rises, do college towns see a higher increase in home sale prices from the previous year than do non-college towns.

In years when the average sale price of single family homes falls, do college towns see a lower decreases in home sale prices from the previous year than do non-college towns.

Is the presence of a college town a stronger predictor of higher increase in home values than is population size.

Are the above values consitnent in the months since March 2020 with the historic values since 1996.
